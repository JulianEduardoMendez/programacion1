```python
usuarios = {
    "ANA": "1234",
    "JUAN": "abcd"
}

try:
    usuario = input("Usuario: ").upper()

    if usuario not in usuarios:
        raise KeyError("El usuario no existe")

    elif usuarios[usuario] != input("Contraseña: "):
      raise ValueError("Contraseña incorrecta")

    else:
        print("Bienvenido,", usuario)

except KeyError as e:
    print("Error:", e)

except ValueError as e:
    print("Error:", e)
```

La diferencia entre los códigos es que en el primero código, el error de usuario no se muestra hasta que no se tenga la contraseña, mientras que el otro código apenas note que el usuario es inválido, suelta el error.

```python
try:
  num1 = float(input("Número 1: "))
  num2 = float(input("Número 2: "))
  op = input("Operación (+, -, *, /, %): ")

  if op == "+":
    resultado = num1 + num2
  elif op == "-":
    resultado = num1 - num2
  elif op == "*":
    resultado = num1 * num2
  elif op == "/":
    resultado = num1 / num2
  elif op == "%":
    resultado = num1 % num2
  else:
    raise ValueError("Operación no válida")

  print("Resultado:", resultado)

except ValueError as e:
  print("Error de valor:", e)
except ZeroDivisionError as e:
  print("No se puede dividir entre cero", e)
except TypeError as e:
  print("Error de tipo:", e)
```

En el primer código, el float se encarga de convertir lo que el usuario escriba a un dato tipo float para poder hacer las operaciones numéricas, si no escribimos un número, nos soltará un TypeError porque lo que se escribió, probablemente no se pueda convertir a un float. Además, si el segundo número es 0, entonces soltará un error de division entre cero, pues no está definida

En el segundo código, se eliminó el float para que el programa interprete los números ingresados como cadenas, por lo que solo estaría definida la suma en este cas, lo demás nos suelta un error

```python
try:
  productos = {
      "pan": 1.5,
      "leche": 2.0,
      "huevos": 3.0
  }

  print("Productos disponibles:", productos)

  nombre = input("¿Qué producto quieres?: ").lower()
  cantidad = int(input("Cantidad: "))

  if nombre not in productos:
    raise KeyError("Producto no disponible")
  if cantidad <= 0:
    raise ValueError("La cantidad debe ser mayor a 0")

  total = productos[nombre] * cantidad
  print("Total a pagar:", total)
except ValueError as e:
  print("Error de valor:", e)
except KeyError as e:
  print("Producto no disponible:", e)
```

```python
import re
```

```python
texto = "Python es un lenguaje"
resultado = re.search("Python", texto)
print(resultado)
```

```python
resultado1 = re.search("x", texto)
print(resultado1)
```

El método search busca la subcadena del primer argumento dentro de la cadena del segundo argumento y me dice en que posiciones está

```python
texto = "Tengo 2 perros y 3 gatos"
resultado3 = re.findall(r"\d", texto)
print(resultado3)
```

El método findall permite encontrar números dentro de cadenas

```python
texto = "Tengo 23 perros y 3 gatos"
resultado3 = re.findall(r"\d+", texto)
print(resultado3)
```

Si escribimos + ahora sí capta los números de dos digitos

```python
texto = "Python es divertido"
resultado4 = re.findall(r"\w+", texto)
print(resultado4)
```

El método también puede escribir en listas las palabras en una lista

```python
texto = "Me gusta Java"
nuevo = re.sub("Java", "Python", texto)
print(nuevo)
```

El método sub sustituye la cadena del primer argumento por la del segundo argumento en la cadena del tercer argumento

```python
texto = "rojo,verde,azul"
resultado5 = re.split(",", texto)
print(resultado5)
```
