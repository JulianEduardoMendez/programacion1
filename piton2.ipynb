```python
a, b = 0, 1
while(b) < 100: #escribir ":" para limite
  print(b)
  a = b
  b = a+b #el valor de b está actualizado porque se dá el espacio de hacerlo
```

```python
a, b = 0, 1
while(b) < 100:
  print(b)
  a, b = b, a+b #toma el valor de b que no está actualizado porque no se da el espacio
```

```python
n = int(input("inserte un numero de 1 a 10"))
if n == 1 or n == 0:
    print("El número es ", n)
else:
  for x in range(2, n):
    if n % x == 0:
        print(n, 'es igual a', x, '*', n // x)
        break
  else: # Este else corresponde al for, no al if
      print(n, 'es un número primo')
```

```python
tupla = (10, 20, 30, 40, 50)
try:
    index = int(input("Ingresa un índice: "))
    print("El valor es:", tupla[index])
except IndexError: # Indice fuera de rango
    print("Índice fuera de rango")
except ValueError: # Cuando ingresamos un valor no permitido
    print("Debes ingresar un número entero")
```

```python
clave_correcta = "python123"
intentos = 3
while intentos > 0:
  intento = input("ingresa la contraseña: ")
  if intento == clave_correcta:
    print("Acceso concedido")
    break
  else:
    intentos -= 1
    print("intentos restantes:", intentos)
else:
  print("cuenta bloqueada")
```
