```python
# Ejemplo de conversión de distintos tipos a False

print(0, "->", bool(0))
print(0.0, "->", bool(0.0))
print("", "->", bool(""))
print(list(), "->", bool(list()))
print(tuple(), "->", bool(tuple()))
print(set(), "->", bool(set()))
print(dict(), "->", bool(dict()))
print(None, "->", bool(None))
```

```python
# Ejemplo de conversión de distintos tipos a True

print(2, "->", bool(2))
print("hola", "->", bool("hola"))
print("Cadena", "->", bool("Cadena"))
print(list((1, 2)), "->", bool(list((1, 2))))
print(tuple((1, 2)), "->", bool(tuple((1, 2))))
print(set((1, 2)), "->", bool(set((1, 2))))
print(dict([(1,2)]), "->", bool(dict([(1,1)])))
```

```python
#Ejemplo if-elif-else

nota = float(input("Introduzca su nota: "))
if 0 <= nota and nota < 3:
  print("Estás suspenso, PFU")
elif 3 <= nota and nota < 3.2:
  print("Raspando")
elif 3.2 <= nota and nota < 4.5:
  print("Super")
elif 4.5 <= nota and nota <= 5:
  print("Cum Laude")
else:
  print("Nota fuera de rango")
```

```python
# for sobre estructuras de datos

lista = list(("a", "b", "c"))
tupla = tuple(("a", "b", "c"))
conjunto = set(("a", "b", "c"))
diccionario = dict(((2,"a"), (1, "b"), (0, "c")))

for i in lista:
  print(i)
for i in tupla:
  print(i)
for i in conjunto:
  print(i)
for i in diccionario:
  print(i)
```

```python
# for para visualizar el efecto de las sentencias break y continue

for numero in range(8):
  print(numero) #Nota: no imprime el valor designado en el argumento de range
```

```python
# Visualización del efecto de la sentencia break

for numero in range(8):
  if numero == 6:
    break
  print(numero)
```

```python
# Visualización del efecto de la sentencia continue

for numero in range(8):
  if numero == 6:
    continue
  print(numero)
```
