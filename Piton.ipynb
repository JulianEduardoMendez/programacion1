Control de flujo

```python
a, b = 0, 1
while(b) < 100: #escribir ":" para limite
  print(b)
  a = b
  b = a+b #el valor de b está actualizado porque se dá el espacio de hacerlo
```

```python
a, b = 0, 1
while(b) < 100:
  print(b)
  a, b = b, a+b #toma el valor de b que no está actualizado porque no se da el espacio
```

```python
a, b = 0, 1
for b in range(21):
  print(b)
  a, b =b, a+b
```

```python
a, b = 0, 1
for b in [5, 8, 9]:
  print(b)
  a, b =b, a+b
```

```python
x = int(input("ingresa un entero, por favor: "))
if x < 0:
  x = 0
  print("Negativo cambiado a cero")
elif x == 0:
  print("Cero")
elif x == 1:
  print("uno")
else :
  print("Número mayor a uno")
```
