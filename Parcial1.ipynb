El método .symetric_difference() agrupa en un conjunto todos los valores que no se encuentren en ambos conjuntos al mismo tiempo, es decir, toma todos los valores de la union de los conjuntos A y B, pero desprecia los que se encuentran en la union. en este caso, crea el conjunto conformado por 3, 4, 5, 6 y 7

Si cumple la condición, entonces para cada elemento n dentro del rango de 0 y el dato hace lo siguiente: si el elemento n es 1 o 0, solo imprime que el numero es 0 y 1, si n no cumple la condición, entonces para cada elemento x dentro del rango de 2 y n evalúa lo siguiente: si x es divisor exacto de n, entonces imprima el mensaje n es igual a la factorización básica de n. Ahí rompe el programa. Si no es cumple la condición anterior, pues imprime el mensaje de que el elemento n es un número primo

**Segunda parte del código**: si no se cumple la primera parte del código, entonces evalúa si el dato ingresado es un elemento tipo string. Si es verdadero, entonces imprima el mensaje "el valor ingresado no es un número"

**Tercera parte del código**: Si no se cumple las dos anteriores, entonces evalúa si el objeto es de tipo float. Si es verdadeo, entonces imprime el mensaje "Por favor, ingresar un dato tipo int, no foat"

```python
def contador_vocales(palabra):

  #Primera parte del código
  vocales = "aeiou"
  vocales1 = vocales.upper()
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  consonantes1 = consonantes.upper()
  numeros = "0123456789"
  espacio = " "
  contador = 0
  contador1 = 0
  contador2 = 0
  contador3 = 0
  contador4 = 0
  contador5 = 0
  contador6 = 0
  #segunda parte del código
  for letra in palabra:
    if letra in vocales:
      contador += 1
    elif letra in vocales1:
      contador1 += 1
    elif letra in consonantes:
      contador2 += 1
    elif letra in consonantes1:
      contador3 += 1
    elif letra in numeros :
      contador4 += 1
    elif letra in espacio:
      contador5 += 1
    else:
      contador6 += 1
  #tercera parte del código
  return({"Cantidad de vocales minúsculas:": contador,
  "Cantidad de vocales mayúsculas:": contador1,
  "Cantidad de consonantes minúsculas:": contador2,
  "Cantidad de consonantes mayúsculas": contador3,
  "Cantidad de números": contador4,
  "Cantidad de espacios": contador5,
  "Cantidad de otros caracteres": contador6})
#cuarta parte del çodigo
cadena = input("Ingrese una cadena de texto: ")
contador_vocales(cadena)
```

**Primera parte del código**: Primero se define las cadenas que se usarán para decir quien es una vocal minúscula, vocal mayúscula, consonante minúscula, consonante mayúscula, número y espacio. Después se definen los contadores para contar cada una de ellas

**Segunda parte del código**: para cada index de la cadena ingresada, evalúa lo siguiente: Si se encuentra en alguna de las variables previamente definidas en la primera parte del código, entonces va aumentando y actualizando el valor de su contador correspondiente. Si el caracter no pertenece a ninguna variable, entonces aumenta y actualiza en 1 su contado por cada item que no se encuentre en las variables

**Tercera parte del código**: Al final, la funcion devolverá un diccionario, siendo las claves el texto respectivo a sus variables, y el valor son los contadores correspondientes a su categoría

**Cuarta parte del código**: Al final, en el código, se usa la funcion input() para que el usuario pueda ingresar una cadena y después esa cadena sea evaluada por la función que definimos en la primera, segunda y tercera parte del código

```python
cadena = input("Ingrese una cadena de texto: ")
contador_vocales(cadena)
```
