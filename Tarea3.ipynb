```python
def pedir_complejo():
  while True:
    try:
      a = list(map(float, input("Ingrese la parte real y la parte imaginaria separadas por ';' ").split(";")))
      if len(a) !=2:
        print("Error: Debe ingresar dos valores separados por ';'.")
      else:
        z = complex(a[0], a[1])
        return z
    except ValueError:
      print("Error: Debe ingresar dos valores numericos separados por ';'.")

def suma(z1, z2):
    resultado = z1 + z2
    return resultado

def multiplicacion(z1, z2):
    resultado = z1 * z2
    return resultado

def raiz_n(z):
  while True:
    try:
      n = int(input("Ingrese el valor de n: "))
      if n == 0:
        print("Error: n no puede ser 0.")
      else:
        resultado = z ** (1/n)
        return resultado
    except ValueError:
      print("Error: n debe ser un numero entero")

print('''MENU DE OPCIONES
1. Sumar numeros complejos
2. Multiplicar numeros complejos
3. Raiz n-esima de un numero complejo
4. Salir''')
while True:
  try:
    opcion = int(input("\nIngresa una opcion: "))

    if opcion == 1:
        z1 = pedir_complejo()
        z2 = pedir_complejo()
        r = suma(z1, z2)
        print(f"Resultado: {r}")


    elif opcion == 2:
        z1 = pedir_complejo()
        z2 = pedir_complejo()
        r = multiplicacion(z1, z2)
        print(f"Resultado: {r}")

    elif opcion == 3:
        z = pedir_complejo()
        r = raiz_n(z)
        print(f"Resultado: {r}")

    elif opcion == 4:
        print("Saliendo del programa...")
        break

    else:
        print("Opcion invalida, intente nuevamente.")
  except ValueError:
    print("Error: Debe ingresar un numero entero.")
```

```python
def pedir_vector():
  a = list(map(float, input("Ingresa un vector separado por espacios: ").split()))
  return a

def suma_vectores(v1, v2):
  if len(v1) == len(v2):
      return [v1[i] + v2[i] for i in range(len(v1))]
  else:
    return "Los vectores deben tener la misma dimensión"

def producto_punto(v1, v2):
  if len(v1) == len(v2):
      return sum([v1[i] * v2[i] for i in range(len(v1))])
  else:
    return "Los vectores deben tener la misma dimensión"

def producto_cruz(v1, v2):
  if len(v1) == 3 and len(v2) == 3:
    x = v1[1] * v2[2] - v1[2] * v2[1]
    y = v1[2] * v2[0] - v2[2] * v1[0]
    z = v1[0] * v2[1] - v1[1] * v2[0]
    return [x, y, z]
  else:
    return "El producto cruz solo está definido para vectores 3D"

def producto_hadamard(v1, v2):
  if len(v1) == len(v2):
    return [v1[i] * v2[i] for i in range(len(v1))]
  else:
    return "Los vectores deben tener la misma dimensión"

print('''Hola, tenemos las siguientes opciones
1) Suma de vectores
2) Producto punto entre vectores
3) Producto cruz entre vectores
4) Producto Hadamard entre vectores
5) Salir''')
while True:
  try:
    respuesta = int(input("\nselecciona una opcion: "))
    if respuesta == 1:
      vector1 = pedir_vector()
      vector2 = pedir_vector()
      print(f"La suma de vectores es {suma_vectores(vector1, vector2)}")
      continue
    elif respuesta == 2:
      vector1 = pedir_vector()
      vector2 = pedir_vector()
      print(producto_punto(vector1, vector2))
      continue
    elif respuesta == 3:
      vector1 = pedir_vector()
      vector2 = pedir_vector()
      print(producto_cruz(vector1, vector2))
      continue
    elif respuesta == 4:
      vector1 = pedir_vector()
      vector2 = pedir_vector()
      print(producto_hadamard(vector1, vector2))
      continue
    elif respuesta == 5:
      print("Gracias por usar el programa")
      break
    else:
      print("Debes elegir una opción válida")
      continue
  except ValueError:
    print("Debes ingresar un número")
```

```python
def matriz():
  while True:
    b = input("Ingrese las filas separadas por ';': ").split(";")
    matriz = []
    value = True
    try:
      for i in range(len(b)):
        c = b[i].split()
        if len(c) == 0:
          print("Error: debes digitar algun numero")
          value = False
          break
        else:
          a = list(map(float,c))
          matriz.append(a)
      if value == False:
        continue

      for j in range(len(matriz)):
        if len(matriz[j]) != len(matriz[0]):
          value = False
          break
      if value == True:
        return matriz
      else:
        print("Error: La cantidad de columnas debe ser igual para todas las filas")
    except ValueError:
      print("Error: Debes poner numeros")

def suma(a, b):

  if len(a) != len(b) or len(a[0]) != len(b[0]):
    print("Error: Las matrices deben tener las mismas dimensiones para sumarse.")
  else:
    result = []
    for i in range(len(a)):
      fila = []
      for j in range(len(a[0])):
        fila.append(a[i][j] + b[i][j])
      result.append(fila)

    print("Resultado:")
    for fila in result:
      print(fila)

def multiplicacion(a, b):

  if len(a[0]) != len(b):
    print("Error: la primera matriz debe tener igual cantidad de columnas que filas en la segunda matriz")

  else:
    resultado = []

    for i in range(len(a)):
      fila = []
      for j in range(len(b[0])):
        suma = 0
        for k in range(len(a[0])):
          suma += a[i][k] * b[k][j]

        fila.append(suma)
      resultado.append(fila)

    print("Resultado:")
    for fila in resultado:
      print(fila)

def p_h(a, b):

  if len(a) != len(b) or len(a[0]) != len(b[0]):
    print("Error: Las matrices deben tener las mismas dimensiones")

  else:
    result = []

    for i in range(len(a)):
      fila = []
      for j in range(len(a[0])):
        fila.append(a[i][j] * b[i][j])
      result.append(fila)

    print("\nResultado:")
    for fila in result:
      print(fila)

def p_k(a, b):

  result = []

  for i in range(len(a)):
    for j in range(len(b)):
      fila = []

      for k in range(len(a[0])):
        for l in range(len(b[0])):
          fila.append(a[i][k] * b[j][l])
      result.append(fila)

  print("Resultado:")
  for fila in result:
    print(fila)

print('''Bienvenido a la calculadora de matrices
1) Sumar Matrices
2) Multiplicar Matrices
3) Producto Hadamard
4) Producto Kronecker
5) Salir''')
while True:
  try:
    opcion = int(input("\n¿Que quiere hacer?: "))
    if opcion == 1:
      print("Las matrices deben tener la misma dimension")
      a = matriz()
      b = matriz()
      suma(a, b)
    elif opcion == 2:
      print("La cantidad de columnas de la primera matriz deben ser la misma cantidad de filas en la segunda")
      a = matriz()
      b = matriz()
      multiplicacion(a, b)
    elif opcion == 3:
      print("Las matrices deben tener la misma dimension")
      a = matriz()
      b = matriz()
      p_h(a, b)
    elif opcion == 4:
      a = matriz()
      b = matriz()
      p_k(a, b)
    elif opcion == 5:
      print("\nGracias por usar la calculadora de matrices")
      break
    else:
      print("Error: Ingrese una opcion valida")
  except ValueError:
    print("Error: Debe ingresar un numero")
```

```python
def matriz():
  while True:
    b = input("Ingrese las filas separadas por ';': ").split(";")
    matriz = []
    value = True
    try:
      for i in range(len(b)):
        c = b[i].split()
        if len(c) == 0:
          print("Error: debes digitar algun numero")
          value = False
          break
        else:
          a = list(map(float,c))
          matriz.append(a)
      if value == False:
        continue

      for j in range(len(matriz)):
        if len(matriz[j]) != len(matriz[0]):
          value = False
          break
      if value == True:
        return matriz
      else:
        print("Error: La cantidad de columnas debe ser igual para todas las filas")
    except ValueError:
      print("Error: Solo puede poner numeros")

def Transpuesta(a):

  matriz_t = []

  for j in range(len(a[0])):
    fila = []
    for i in range(len(a)):
      fila.append(a[i][j])
    matriz_t.append(fila)

  for fila in matriz_t:
    print(fila)

def submatriz(a, fila, colu):
  matriz = []

  for i in range(len(a)):
    if i == fila:
      continue

    filas = []
    for j in range(len(a[0])):
      if j != colu:
        filas.append(a[i][j])
    matriz.append(filas)

  return matriz

def determinante(a):
  if len(a) == 1:
    return a[0][0]
  if len(a) == 2:
    return a[0][0]*a[1][1]-a[0][1]*a[1][0]

  det = 0
  for i in range(len(a)):
    det += ((-1)**i)*a[0][i]*determinante(submatriz(a,0,i))
  return det

def m_cofactores(a):
  if len(a) == len(a[0]):
    cofactores = []

    for i in range(len(a)):
      fila = []
      for j in range(len(a[0])):
        fila.append(((-1)**(i+j))*determinante(submatriz(a, i, j)))
      cofactores.append(fila)

    for filas in cofactores:
      print(filas)
  else:
    print("Error: La matriz debe ser cuadrada")

def vec_fil(a):
  vector = []
  for i in range(len(a)):
    for j in range(len(a[0])):
      vector.append(a[i][j])
  print(vector)

print("Bienvenido a la transformadora de Matrices")
print('''1) Transponer una matriz
2) Determinar la matriz de cofactores
3) Convertir una matriz en un vector fila
4) Salir''')
while True:
  try:
    opcion = int(input("\n¿Que quieres hacer?: "))
    if opcion == 1:
      a = matriz()
      Transpuesta(a)
    elif opcion == 2:
      print("Recuerda que la matriz debe ser cuadrada")
      a = matriz()
      m_cofactores(a)
    elif opcion == 3:
      a = matriz()
      vec_fil(a)
    elif opcion == 4:
      print("Hasta luego :)")
      break
    else:
      print("Error: numero fuera de opciones")

  except ValueError:
    print("Error: digita un numero de los indicados")
```
