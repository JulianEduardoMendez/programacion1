```python
PruebaA = np.random.normal(loc = 7.2, scale = 0.4, size = 53)
PruebaB = np.random.gamma(shape = 306, scale = 0.0229, size = 53)
```

```python
df0 = pd.DataFrame({
    "A": PruebaA,
    "B": PruebaB
})
```

```python
df0.describe()
```

```python
df0.median()
```

La media del A es 7.2 y del B es 6.91 aprox

La desviacion estandar de A es 0.43 y de B es 0.35 aprox

Un cuarto de los datos son inferiores a 6.9 en A y a 6.68 en B

Un 75% de los datos son inferiores a 7.49 en A y a 7.13 en B

El valor máximo de A es 8.23 y de B es 7.63 aprox

El valor mínimo de A es 6.38 y de B es 5.98

La mediana del A es 7.21 y del B es 6.95

```python
from scipy import stats
```

```python
t_stat, p_val = stats.ttest_ind(df0["A"], df0["B"])

print("Estadístico t:", t_stat)
print("Valor p:", p_val)
```

No se cumple el supuesto de normalidad con una confianza del 95%, pues el valor p es inferiror a 0.05

```python
stats.ttest_rel(df0["A"], df0["B"])
```

Tambien se rechaza que el puntaje medio en ambas encuestas sea igual con una confianza del 95%, pues el valor p es menor a 0.05

```python
stats.kendalltau(df0["A"], df0["B"])
```

No se rechaza que los puntajes de ambas encuestas no están correlacionado con una confianza del 95%, pues el valor p es mayor a 0.05

```python
stats.spearmanr(df0["A"], df0["B"])
```

No se rechaza con una confianza del 95% que los puntajes de ambas encuestas no estan correlacionados
